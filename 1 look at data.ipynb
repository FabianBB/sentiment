```python
# Convert 'Date' to datetime and set it as index to facilitate time series operations
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Calculate daily returns for the 'Close' prices
data['Returns'] = data['Close'].pct_change()

# Create lagged sentiment features: 1-day, 7-day, and 30-day lags
lag_days = [1, 7, 30]
sentiment_columns = ['Bullish_sentiment', 'Neutral_sentiment', 'Bearish_sentiment']

for column in sentiment_columns:
    for lag in lag_days:
        data[f'{column}_lag{lag}'] = data[column].shift(lag)

# Drop NaN values from the data after shifting (which introduces NaNs)
data.dropna(inplace=True)

# Show a snippet of the new dataset with the lagged features and daily returns
data[['Returns', 'Bullish_sentiment_lag1', 'Neutral_sentiment_lag1', 'Bearish_sentiment_lag1',
      'Bullish_sentiment_lag7', 'Neutral_sentiment_lag7', 'Bearish_sentiment_lag7',
      'Bullish_sentiment_lag30', 'Neutral_sentiment_lag30', 'Bearish_sentiment_lag30']].head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Selecting features and target for the regression model
features = ['Bullish_sentiment_lag1', 'Neutral_sentiment_lag1', 'Bearish_sentiment_lag1']
target = 'Returns'

# Prepare the data for training and testing
X = data[features]
y = data[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Output the coefficients and the model performance
coefficients = pd.Series(model.coef_, index=features)
mse, r2, coefficients
```

```python
# Adding rolling averages for 'Close' prices over 7 and 30 days
data['Close_roll_avg7'] = data['Close'].rolling(window=7).mean()
data['Close_roll_avg30'] = data['Close'].rolling(window=30).mean()

# Drop NaN values introduced by rolling averages
data.dropna(inplace=True)

# Updated set of features including more lagged sentiment and rolling averages
features_extended = ['Bullish_sentiment_lag1', 'Neutral_sentiment_lag1', 'Bearish_sentiment_lag1',
                     'Bullish_sentiment_lag7', 'Neutral_sentiment_lag7', 'Bearish_sentiment_lag7',
                     'Bullish_sentiment_lag30', 'Neutral_sentiment_lag30', 'Bearish_sentiment_lag30',
                     'Close_roll_avg7', 'Close_roll_avg30']

# Prepare the data for the regression model
X = data[features_extended]
y = data['Returns']

# Create and train the model on the full dataset
model_extended = LinearRegression()
model_extended.fit(X, y)

# Coefficients of the model
coefficients_extended = pd.Series(model_extended.coef_, index=features_extended)

# Predict using the model
y_pred_full = model_extended.predict(X)

# Calculate metrics for the full dataset
mse_full = mean_squared_error(y, y_pred_full)
r2_full = r2_score(y, y_pred_full)

mse_full, r2_full, coefficients_extended
```

```python
# Assuming data already includes 'Returns'
# Create lagged sentiment indicators for various lags
lags = [1, 5, 10, 20]  # You can adjust or extend this list based on your specific analysis needs
sentiment_columns = ['Bullish_sentiment', 'Neutral_sentiment', 'Bearish_sentiment']

# Generate lagged columns
for col in sentiment_columns:
    for lag in lags:
        data[f'{col}_lag{lag}'] = data[col].shift(lag)

# Calculate correlations between lagged sentiments and returns
correlation_results = {}
for col in sentiment_columns:
    correlations = []
    for lag in lags:
        correlation = data['Returns'].corr(data[f'{col}_lag{lag}'])
        correlations.append((f'{col}_lag{lag}', correlation))
    correlation_results[col] = correlations

# Output the correlation results nicer
for key, values in correlation_results.items():
    print(f'Correlations for {key}:')
    for value in values:
        print(f'{value[0]}: {value[1]:.4f}')
    print('\n')
```

```python
# Calculate 30-day rolling volatility of returns
data['Rolling_Volatility_30'] = data['Returns'].rolling(window=30).std()

# Calculate correlations between lagged sentiments and rolling volatility
volatility_correlation_results = {}
for col in sentiment_columns:
    vol_correlations = []
    for lag in lags:
        correlation = data['Rolling_Volatility_30'].corr(data[f'{col}_lag{lag}'])
        vol_correlations.append((f'{col}_lag{lag}', correlation))
    volatility_correlation_results[col] = vol_correlations

for key, values in volatility_correlation_results.items():
    print(f'Correlations with Rolling Volatility for {key}:')
    for value in values:
        print(f'{value[0]}: {value[1]:.4f}')
    print('\n')
```

```python
# same for 7 day rolling volatility
data['Rolling_Volatility_7'] = data['Returns'].rolling(window=7).std()

# Calculate correlations between lagged sentiments and rolling volatility
volatility_correlation_results = {}
for col in sentiment_columns:
    vol_correlations = []
    for lag in lags:
        correlation = data['Rolling_Volatility_7'].corr(data[f'{col}_lag{lag}'])
        vol_correlations.append((f'{col}_lag{lag}', correlation))
    volatility_correlation_results[col] = vol_correlations
    
for key, values in volatility_correlation_results.items():
    print(f'Correlations with Rolling Volatility for {key}:')
    for value in values:
        print(f'{value[0]}: {value[1]:.4f}')
    print('\n')
```

```python
from statsmodels.tsa.stattools import grangercausalitytests

# Prepare the data for Granger Causality Test: 'Bearish_sentiment' and 'Rolling_Volatility_30'
maxlag = 20
test_data = data[['Rolling_Volatility_30', 'Bearish_sentiment']].dropna()

# Granger Causality Test
gc_results = grangercausalitytests(test_data, maxlag=maxlag, verbose=True)

# Collect the p-values for the test at each lag
gc_p_values = {lag: result[0]['ssr_chi2test'][1] for lag, result in gc_results.items()}
gc_p_values
```

Interpretation:
High Significance at Shorter Lags: The bearish sentiment Granger-causes volatility significantly at shorter lags (1 to 10 days), with the strongest effects seen at the very shortest lags. This suggests that recent bearish sentiments can be a significant predictor of near-term market volatility.
Decreased Influence at Longer Lags: As the lag increases, the predictive power of bearish sentiment on volatility diminishes. This could indicate that the market's response to bearish sentiments is more immediate and does not persist far beyond a 10-day period.
Implications:
Given the significant results at shorter lags, bearish sentiment appears to be a useful predictor of market volatility and should be considered in risk management strategies and when forecasting market dynamics. The rapid decrease in significance with increased lags suggests that bearish sentiment affects the market more acutely, influencing short-term volatility rather than longer-term trends.

```python
import statsmodels.api as sm

# Assuming 'Rolling_Volatility_30' is your dependent variable and already calculated
# Prepare the independent variables (bearish sentiment lags)
# Let's use significant lags from the Granger causality test

# Prepare X matrix with lagged sentiment values (using lags 1 to 10 as example)
X = sm.add_constant(data[['Bearish_sentiment_lag1', 'Bearish_sentiment_lag2', 'Bearish_sentiment_lag3',
                          'Bearish_sentiment_lag4', 'Bearish_sentiment_lag5', 'Bearish_sentiment_lag6',
                          'Bearish_sentiment_lag7', 'Bearish_sentiment_lag8', 'Bearish_sentiment_lag9',
                          'Bearish_sentiment_lag10']].dropna())

# Prepare Y vector for regression
Y = data.loc[X.index, 'Rolling_Volatility_30']  # Make sure Y aligns with X's indexing

# Fit the linear regression model
model = sm.OLS(Y, X).fit()

# Output the results
print(model.summary())
```

Interpretation of the OLS Regression Results
The regression analysis provides a clearer understanding of how lagged bearish sentiment impacts market volatility. Here's a detailed interpretation of the results:

Key Statistics:
R-squared: 0.243 - This indicates that approximately 24.3% of the variability in market volatility is explained by the lagged values of bearish sentiment. This is a significant proportion, suggesting that bearish sentiment is indeed an important factor in predicting market volatility.
Adjusted R-squared: 0.236 - This value is slightly adjusted for the number of predictors and gives a more accurate measure of the model's explanatory power.
F-statistic: 36.62 - The model is statistically significant with a probability (Prob(F-statistic)) of about 1.73e-62, which is extremely low, indicating a very strong model fit.
Coefficients:
Constant: The constant term is not significant (P>|t| = 0.119), indicating no significant intercept adjustment is needed beyond the influence of the predictors.
Bearish_sentiment_lag1 to Bearish_sentiment_lag10: The coefficients associated with these lags tell us the size and direction of the impact that bearish sentiment from 1 to 10 days prior has on current market volatility.
Lag 1 and Lag 10 are particularly notable:
Lag 1: Coefficient = 0.0091, P>|t| = 0.030 - A significant positive effect, suggesting that an increase in bearish sentiment from the previous day slightly increases volatility.
Lag 10: Coefficient = 0.0150, P>|t| = 0.000 - The most significant and largest positive effect, indicating that bearish sentiment from 10 days ago has a more substantial impact on increasing current volatility.
Other Lags: The other lags show positive coefficients but are not statistically significant, indicating weaker or no discernible effects at these specific intervals.

Omnibus, Jarque-Bera Test: These tests assess the normality of the residuals.
