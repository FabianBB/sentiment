```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
# get sp500 data 
sp500 = yf.download('^GSPC', start='2000-01-06')
sp500
```

```python
df = pd.read_csv("sentiment.csv")
df
```

```python
# for all columns in df (except date) rename the column to the column name + _sentiment
for col in df.columns[1:]:
    df.rename(columns={col: col + "_sentiment"}, inplace=True)
```

```python
# convert the date column to datetime for df and sp500
df['Date'] = pd.to_datetime(df['Date'])
sp500['Date'] = pd.to_datetime(sp500.index)
# rename the index to ind
sp500.index.name = 'ind'
```

```python
# merge the two dataframes on the date column
merged = pd.merge(sp500, df, on='Date')
```

```python
merged
```

```python
# make date the first column
a = merged.pop('Date')
merged.insert(0, 'Date', a)
merged
```
